```python
# if_shuffle
if_shuffle = [False,True]
if_shuffle_flag = ['No_Shuffle','Shuffle']
num_shuffle_choice = len(if_shuffle)
#dff0_thresh
dff0_thresh = 1
# cv fold
cv_fold = 5
list_channel_selected = [0,1,2]# 0-both channels; 1-G7f; 2-NM
num_channel_choice = len(list_channel_selected)
list_odor_choice = [0] # 0-All; 1-no oct; 2-no mch; 3-no ea
num_odor_choice = len(list_odor_choice)
list_thresh_pca_exp_var_2 = np.arange(0,1,0.005)# pca theshold-2
num_thresh_points = len(list_thresh_pca_exp_var_2)


## read data
data_path = '../../../results/3.odor_random_90_times_OCT_MCH_EA_new/nsyb-G7f-rAch1h'
result_path = '../results/Ach-final-test'
file_name = ['dff0_0-80_down2_C2','dff0_0-80_down2_C3']
atlas_path = data_path
atlas_name = 'align_to_atlas/Transformed_atlas.tif'
atlas_eroded_name = 'align_to_atlas/Transformed_atlas_eroded_r5.tif'
file_folders = file_folder_list_1
num_fly = len(file_folders)
num_channel = 2
atlas_z_range = range(13,38)
#####
fly_selected = 1
print(file_folders)
print(file_folders[fly_selected])
if_plot_figure = False
if_save_figure = True
region_c1 = np.array([4,7,12,16,17,18,19,23,26,27,28,29,34,39,55,59,63,64,65,66,72,73,74,79,84])
# 'NO','LH_R','ICL_R','CRE_R','MBPED_R','MBVL_R','MBML_R','EB','FB','SLP_R','SIP_R','SMP_R',
# 'AOTU_R','SCL-R','LH_L','ICL_L','CRE_L','MBPED_L','MBVL_L','MBML_L','SLP_L','SIP_L','SMP_L',
# 'AOTU_L','SCL_L'


## parameters for PCA
pca_tp_range = range(1,15)
## parameters for LDA and SVM
lda_tp_selected = 3
svm_tp_range = range(1,15)
```

```python
## only 1 fly, 1 condition
# load data
file_folder_1 = file_folder_list_1[fly_selected]
path_file_0 = data_path + '/' + file_folder_1 + '/' + file_name[0] + '.npy'
path_file_1 = data_path + '/' + file_folder_1 + '/' + file_name[1] + '.npy'
data_1 = np.load(path_file_0)
data_1 = np.transpose(data_1,[3,4,2,1,0])
data_2 = np.load(path_file_1)
data_2 = np.transpose(data_2,[3,4,2,1,0])
size_x = np.size(data_1,0)
size_y = np.size(data_1,1)
size_z = np.size(data_1,2)
num_tp = np.size(data_1,3)
num_trial = np.size(data_1,4)
data = np.concatenate((data_1,data_2),axis = 4)
del data_1,data_2
data = data.reshape(size_x,size_y,size_z,num_tp,num_trial,num_channel,order = 'F')
print('load data done! size:')
print(np.shape(data))

# load atlas
file_folder_2 = file_folder_1[:-5]
atlas= tf.imread(atlas_path + '/' + file_folder_2 + '/' + atlas_name)
atlas = np.transpose(atlas,[1,2,0])
atlas = atlas[:,:,atlas_z_range]
print('load atlas done! size:')
print(np.shape(atlas))
atlas_eroded= tf.imread(atlas_path + '/' + file_folder_2 + '/' + atlas_eroded_name)
atlas_eroded = np.transpose(atlas_eroded,[1,2,0])
atlas_eroded = atlas_eroded[:,:,atlas_z_range]
print('load atlas_eroded done! size:')
print(np.shape(atlas_eroded))
```

```python
## consider all variables
list_list_accuracy_br = np.zeros((num_channel_choice,num_odor_choice,num_thresh_points,cv_fold,num_shuffle_choice))
list_list_f1_weighted_br = np.zeros((num_channel_choice,num_odor_choice,num_thresh_points,cv_fold,num_shuffle_choice))
list_list_AUC_weighted_br = np.zeros((num_channel_choice,num_odor_choice,num_thresh_points,cv_fold,num_shuffle_choice))
list_list_accuracy_svm_br = np.zeros((num_channel_choice,num_odor_choice,num_thresh_points,cv_fold,num_shuffle_choice))

result_each_fly_path = result_path + '/' + file_folder_list_2[fly_selected] + '/' + 'mean_region_pca_DEEPCAD_formal'
folder = os.path.exists(result_each_fly_path)
if not folder:
    os.makedirs(result_each_fly_path)

# processing-brain region
for j,channel_selected in enumerate(list_channel_selected):
    for k,odor_choice in enumerate(list_odor_choice):
        print('Brain Region: '+'channel '+str(j)+',odor choice '+str(odor_choice))
        for m in range(num_shuffle_choice):
            result_each_fly_path_1 = os.path.abspath(result_each_fly_path + '/' + if_shuffle_flag[m] + '/' + 'C_' + str(channel_selected)\
                                     + '_odor_choice_'+ str(odor_choice))
            if if_save_figure:
                folder = os.path.exists(result_each_fly_path_1)
                if not folder:
                    os.makedirs(result_each_fly_path_1)
            data_region,stim_2,num_tp,num_trial = produce_brain_region(data,channel_selected,atlas,region_c1,
                            region_c2,region_plus,stim,odor_choice,if_shuffle[m],dff0_thresh,pca_tp_range)
            for q,thresh_pca_exp_var_2 in enumerate(list_thresh_pca_exp_var_2):
                print('thresh:'+str(thresh_pca_exp_var_2))
                data_region,list_accuracy,list_f1_weighted,list_AUC_weighted,list_accuracy_svm = odor_classification_brain_region(data_region,
                                        stim_2,num_tp,
                                        num_trial,thresh_pca_exp_var_2,odor_choice,cv_fold,
                                        pca_tp_range,lda_tp_selected,svm_tp_range,
                                        if_plot_figure,if_save_figure,result_each_fly_path_1)
                list_list_accuracy_br[j,k,q,:,m] = np.array(list_accuracy) 
                list_list_f1_weighted_br[j,k,q,:,m] = np.array(list_f1_weighted)
                list_list_AUC_weighted_br[j,k,q,:,m] = np.array(list_AUC_weighted)
                list_list_accuracy_svm_br[j,k,q,:,m] = np.array(list_accuracy_svm)
                print('Mean accuracy:'+str(np.mean(list_accuracy)))
    

np.save(result_each_fly_path+'/'+'list_list_accuracy_br'+'.npy',list_list_accuracy_br)
np.save(result_each_fly_path+'/'+'list_list_f1_weighted_br'+'.npy',list_list_f1_weighted_br)
np.save(result_each_fly_path+'/'+'list_list_AUC_weighted_br'+'.npy',list_list_AUC_weighted_br)
np.save(result_each_fly_path+'/'+'list_list_accuracy_svm_br'+'.npy',list_list_accuracy_svm_br)
```
